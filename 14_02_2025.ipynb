```python
file_path = "/content/Movie.csv"
df = pd.read_csv(file_path)
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
len(df.movie.unique())
```

```python
user_movies_df = df.pivot(index='userId',
                                 columns='movie',
                                 values='rating').reset_index(drop = True)
```

```python
user_movies_df
```

```python
user_movies_df.index = df.userId.unique()
user_movies_df
```

```python
user_movies_df.isnull().sum()
```

```python
numeric_cols = user_movies_df.select_dtypes(include=['number']).columns
user_movies_df[numeric_cols] = user_movies_df[numeric_cols].fillna(user_movies_df[numeric_cols].median())
```

```python
user_movies_df
```

```python
from scipy.spatial.distance import correlation
vec1 = [5, 4, 0, 0]
vec2 = [4, 5, 3, 0]
correlation(vec1, vec2)
```

```python
user_sim = 1 - pairwise_distances(user_movies_df.values, metric='cosine')
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
 user_sim_df.index = df.userId.unique()
 user_sim_df
```
